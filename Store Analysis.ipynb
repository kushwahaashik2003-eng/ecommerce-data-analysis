```python
ax=sns.countplot(data=df, x='Status')
for container in ax.containers:
    ax.bar_label(container)
plt.title("Status")
plt.show()
```

```python
sns.countplot(data=df, x="Channel ",palette='bright',hue='Channel ')
plt.title("Total Count By Plateform")
plt.show()
```

```python
df.groupby("Channel ")['Amount'].sum().plot.pie(
    autopct='%1.1f%%',
    figsize=(8,8),
    # wedgeprops={'width': 0.4},   #  This creates the donut

    ylabel="")
plt.title("Total Revenue Perchentage by Channel")
plt.show()
```

```python
df.groupby("Category")["Amount"].sum().plot.pie(
    autopct='%1.1f%%',
    figsize=(10,10),
    wedgeprops={'width':0.4},
    pctdistance=1.2,
    fontsize=10,
    labeldistance=1.3,
    xlabel="")
plt.title("Total Revenue by Category")
plt.show()
```

```python
sns.countplot(data=df,x="Gender",hue='Gender')
plt.title("Total Orders By Gender")
plt.show()
```

```python

total_amount=df.groupby('Gender')['Amount'].sum().reset_index()
sns.barplot(data=total_amount ,x="Gender", y='Amount',hue='Gender')
plt.title(" Amount Spend by Gender")
plt.ticklabel_format(style='plain',axis='y')

plt.show()
```

```python
df.head(1)
```

```python
bins = [0,25,35,45,55,65,200]
labels = ['18-25','26-35','36-45','46-55','56-65','65+']
df['Age_group'] =pd.cut(df['Age'],bins=bins, labels=labels)

age_sales=df.groupby('Age_group')['Amount'].sum().reset_index()

sns.barplot(data=age_sales, x="Age_group",y='Amount',hue='Age_group')
plt.ticklabel_format(style='plain',axis='y')
plt.title("Amount Spend By Age Group")
plt.ylabel("Total Amount")
plt.xlabel("Age Group")
plt.show()
```

```python
state_sales=df.groupby('ship-state')['Amount'].sum().reset_index().sort_values('Amount',ascending=False).head(5)
plt.figure(figsize=(8,5))
sns.barplot(data=state_sales,x='ship-state',y='Amount',hue='ship-state')
plt.ticklabel_format(style='plain',axis='y')
plt.title("Top 5 State Wise Revenue")
plt.show()
```

```python
city_wise=df.groupby('ship-city')['Amount'].sum().reset_index().sort_values('Amount', ascending=False).head(5)
sns.barplot(data=city_wise, x='ship-city',y='Amount',hue='ship-city')
plt.title("Top 5 Cities by revenue")
plt.ticklabel_format(style='plain',axis='y')
plt.show()
```

```python
channel_sales=df.groupby('Channel ')['Amount'].sum().reset_index().sort_values('Amount',ascending=False)
sns.barplot(data=channel_sales, x='Channel ',y='Amount',hue='Channel ')
plt.title("Channel Wise Revenue")
plt.ticklabel_format(style='plain',axis='y')
plt.show()
```

```python
top_customer=df.groupby('Cust ID')['Amount'].sum().reset_index().sort_values('Amount',ascending=False).head(10)
plt.figure(figsize=(8,6))
sns.barplot(data=top_customer,x='Cust ID',y='Amount', hue='Amount',palette='bright')
plt.xticks(rotation=45)
plt.title("Top 10 Customers")
plt.show()
```

```python
sns.countplot(data=df,x='Size',hue='Size',palette='bright')
plt.title("Total Count By Size")

plt.show()
```

```python
size_revenue=df.groupby('Size')['Amount'].sum().reset_index().sort_values('Amount',ascending=False)
sns.barplot(data=size_revenue, x='Size',y='Amount',hue='Size',palette='bright')
plt.ticklabel_format(style='plain', axis='y')
plt.title("Total Revenue By Size")
plt.show()
```

```python
df['Date']=pd.to_datetime(df['Date'])
daily_sales=df.groupby('Date')['Amount'].sum().reset_index()
sns.lineplot(data=daily_sales, x='Date',y='Amount')
plt.title("Daily Sales Trend")
plt.show()
```

```python
pivot=df.pivot_table(values='Amount',index='Gender',columns='Category',aggfunc='sum')
plt.figure(figsize=(8,6))
sns.heatmap(pivot, annot=True,fmt='.0f')
plt.title("Gender Vs Category Heatma[p")
plt.tight_layout()
plt.show()
```
